## Plots

```python
messages_csv_abs = pvh.plot_texts(df_csv, rescaled=False, setting=tdh.PlotStyle.COUNT, colors=CUSTOM_COLORS1, time_frequency="M")
messages_json_abs = pvh.plot_texts(df_json, rescaled=False, setting=tdh.PlotStyle.COUNT, colors=CUSTOM_COLORS2, time_frequency="M", remove_reactions = False)
```

```python
chars_csv_abs = tdh.plot_texts(df_csv, rescaled=False, setting=tdh.PlotStyle.VOLUME, time_frequency="W",colors=CUSTOM_COLORS1)
chars_json_abs = tdh.plot_texts(df_json, rescaled=False, setting=tdh.PlotStyle.VOLUME, time_frequency="W",colors=CUSTOM_COLORS2, remove_reactions = False)
```

```python
# keyword in these graphs is "haha". First two graphs show the aggregate of the conversations, nothing about sent/received.
# n=5932 for purple, n=3021 for orange
keyword_csv_abs = tdh.plot_texts(df_csv, rescaled=False, setting=tdh.PlotStyle.KEYWORD, time_frequency="M", show_frequency=True, split_by_direction = False, colors=CUSTOM_COLORS1)
keyword_json_abs = tdh.plot_texts(df_json, rescaled=False, setting=tdh.PlotStyle.KEYWORD, time_frequency="M", show_frequency=True, split_by_direction = False, colors=CUSTOM_COLORS2, remove_reactions = False)
```

```python
reaction_csv_abs = tdh.plot_texts(df_csv, rescaled=False, setting=tdh.PlotStyle.REACTIONS, time_frequency="W", colors=CUSTOM_COLORS1, remove_reactions=False, show_frequency=False, split_by_direction = True)
reaction_json_abs = tdh.plot_texts(df_json, rescaled=False, setting=tdh.PlotStyle.REACTIONS, time_frequency="W", colors=CUSTOM_COLORS2, remove_reactions=False, show_frequency=False, split_by_direction = True)
reaction_csv_rsc = tdh.plot_texts(df_csv, rescaled=True, setting=tdh.PlotStyle.REACTIONS, time_frequency="W", colors=CUSTOM_COLORS1, remove_reactions=False, show_frequency=True, split_by_direction = True)
reaction_json_rsc = tdh.plot_texts(df_json, rescaled=True, setting=tdh.PlotStyle.REACTIONS, time_frequency="W", colors=CUSTOM_COLORS2, remove_reactions=False, show_frequency=True, split_by_direction = True)

```
